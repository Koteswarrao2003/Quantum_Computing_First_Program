```python
from qiskit import *
from qiskit.visualization import plot_histogram
%matplotlib inline

circuit=QuantumCircuit(3,3)
circuit.h(0)
circuit.h(1)
circuit.cx(1,2)
circuit.cx(0,1)
circuit.h(0)

c0=circuit.measure(0,0)
c1=circuit.measure(1,1)

if(c1==0 and c0==1):
    circuit.x(2)
elif(c1==1 and c0==0):
    circuit.z(2)
elif(c1==1 and c0==1):
    circuit.x(2)
    circuit.z(2)

from qiskit_aer import Aer
simulator=Aer.get_backend('qasm_simulator')
job=simulator.run(circuit)
result=job.result()
count=result.get_counts()
plot_histogram(count)
```

```python
from qiskit import *
from qiskit.visualization import plot_histogram
%matplotlib inline

circuit=QuantumCircuit(3,3)
circuit.h(0)
circuit.h(1)
circuit.cx(1,2)
circuit.cx(0,1)
circuit.h(0)

circuit.measure(0,0)
circuit.measure(1,1)

with circuit.if_test((circuit.clbits[1], 1)):
    circuit.x(2)

with circuit.if_test((circuit.clbits[0], 1)):
    circuit.z(2)

circuit.measure(2,2)

from qiskit_aer import Aer
simulator=Aer.get_backend('qasm_simulator')
job=simulator.run(circuit, shots=1024)
result=job.result()
count=result.get_counts()
plot_histogram(count)
```
